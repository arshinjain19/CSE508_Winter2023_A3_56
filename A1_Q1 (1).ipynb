```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Wiki-Vote.txt', sep="\t", skiprows=3)
print(data.head())
print(data.tail())
```

```python
source_node = data['# FromNodeId'].tolist()
target_node = data['ToNodeId'].tolist()

# Combine source_node and target_node lists into a single 'total_node' list
# source_node.extend(target_node)
# total_node = source_node
total_node = source_node + target_node
print(total_node)
print(len(total_node))
```

```python
storenode = []
for i in total_node:
    if i not in storenode:
        storenode.append(i)
print(storenode)
print(len(storenode))
```

Q1-a)

```python
# 0 filled adjacency_matrix 
num_nodes = len(storenode)
adj_matrix = np.zeros((num_nodes, num_nodes), dtype=int)
print(adj_matrix)
```

```python
#ADJACENCY MATRIX
#if there is an edge from node 1 to 2, then the adj_matrix[1][2]=1 else it will be 0.
for i in range(len(source_node)):
    row = storenode.index(source_node[i])   #finds the index of the value source_node[i] in the storenode list and assigns it to row. This index represents the row in the adjacency matrix where the edge corresponding to source_node[i] will be added.
    col = storenode.index(target_node[i])
    adj_matrix[row][col] = 1
print(adj_matrix)
```

```python
#EDGE LIST
edge_list = []
for i in range(len(source_node)):
  edge_list.append((source_node[i], target_node[i]))
    
print(edge_list)
print(len(edge_list))
```

 In the code for average in-degree, we iterate over the columns first and then the rows, while in the code for average out-degree, we iterate over the rows first and then the columns. This is because the rows in the adjacency matrix represent the source vertices (outgoing edges) and the columns represent the destination vertices (incoming edges) in a directed graph. 

```python
# 3)Avg In-degree
indeg = 0
for j in range(len(adj_matrix)):
    for i in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            indeg += 1

avg_in_degree = indeg / len(adj_matrix)
print(avg_in_degree)
# matrix: can directly perform operation the graph and it's more efficient in terms of time and space in comparision with edge list.
```

```python
# 4)Avg Out-degree
outdeg = 0
for i in range(len(adj_matrix)):
    for j in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            outdeg += 1

avg_out_degree = outdeg / len(adj_matrix)
print(avg_out_degree)
#The average indegree and outdegree will be the same because the nodes with indegree will get balanced by the nodes with outdegree.
```

```python
# 5)Node with Max In-degree(using adj_matrix)
max_indegree = 0
max_indegree_node = None

for j in range(len(adj_matrix)):
    indeg = 0
    for i in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            indeg += 1
    if indeg >= max_indegree:
        max_indegree = indeg
        max_indegree_node = j

print("Node with Max Indegree:",storenode[max_indegree_node])
print("In-degree Count of Max Indegree Node:", max_indegree)
```

```python
# 6)Node with Max out-degree(using adj_matrix)
max_outdegree = 0
max_outdegree_node = None
for i in range(len(adj_matrix)):
    outdeg = 0
    for j in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            outdeg += 1
    if outdeg >= max_outdegree:
        max_outdegree = outdeg
        max_outdegree_node = i

print("Node with Max Outdegree:", storenode[max_outdegree_node])
print("Outdegree Count of Max Outdegree Node:", max_outdegree)
```

```python
# 7)The density of the network
'''The density of a network typically ranges from 0 to 1.The network density tells us that if the density is 0, then the network has no
edges,indicating very few connections and if the density is 1,then the network is a complete graph,indicating a fully connected network.
Network density is calculated as : Density (D) = E / (N*(N-1)) for directed networks;E = number of edges,N = number of nodes
'''
Density = len(edge_list)/(len(storenode)*(len(storenode) - 1))
print(Density)
```

out-degree

```python
#  plot in-degree, degree distribution of a directed graph
in_degrees = {}

# Calculate the in-degrees for each node
for j in range(len(adj_matrix)):
    for i in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            if j not in in_degrees:
                in_degrees[j] = 0
            in_degrees[j] += 1

# Print the in-degrees dictionary
# print(in_degrees)
```

```python
inDegLabels = in_degrees.keys()
inDegValues = in_degrees.values()
plt.figure(figsize=(20, 10), dpi=60)
plt.title('In-Degree Distribution')
plt.xlabel('Number of Nodes')
plt.ylabel('In-Degree')
plt.scatter(inDegLabels,inDegValues, color = 'g')
plt.show()
```

```python
 #plot out-degree, degree distribution of a directed graph
out_degrees = {}

# Calculate the out-degrees for each node
for i in range(len(adj_matrix)):
    out_degree = 0
    for j in range(len(adj_matrix)):
        if adj_matrix[i][j] == 1:
            out_degree += 1
    out_degrees[i] = out_degree

# Print the out-degrees dictionary
# print(out_degrees)
```

```python
outDegLabels = out_degrees.keys()
outDegValues = out_degrees.values()
plt.figure(figsize=(20, 10), dpi=60)
plt.title('Out-Degree Distribution')
plt.xlabel('Number of Nodes')
plt.ylabel('Out-Degree')
plt.scatter(outDegLabels,outDegValues, color = 'g')
plt.show()
```

```python
#For calculating local clustering coefficient, we iterated over each node of the graph, then found out it’s 
#neighboring nodes and stored in the list using from-ToNodeId dataframe, then iterated over the dataframe while placing each
# of the neighbouring node as ‘# FromNodeId’ one by one, and checking if the ‘ToNodeId’ also lies from the list created earlier, since our dataset is directed in nature we explicitly took connection from a to b, and , b to a as single edge only.
#Then, while iterating over each node, we calculated clustering coefficient as – 
#(number of connection in neighboring nodes)/(number of possible connection between neighboring node)
#Stored node as key and it’s coefficient value in dictionary.
adj_matrix
```

```python
#Local clustering coefficient of each node = number of pairs of neighbors of the node that are connected/ number of pairs of neighbors of the node
cluster_coeff_of_each_node = [] 
for node in range(adj_matrix.shape[0]):
  neighbours = np.where(adj_matrix[node] > 0)[0] # find the indices of all nodes adjacent to the current node
  neighboursNumber = len(neighbours) 
  neighbouringEdges = 0 
  if neighboursNumber > 1: # only calculate clustering coefficient if there are at least two adjacent nodes
    for i in neighbours: 
      for j in neighbours: 
        if i != j and adj_matrix[i][j] > 0: # count the number of edges between different adjacent nodes
          neighbouringEdges = neighbouringEdges + 1
    neighbouringEdges = neighbouringEdges / 2 # divide by 2 since each edge is counted twice
    coeff = 2 * neighbouringEdges / (neighboursNumber * (neighboursNumber - 1)) #clustering coefficient
    cluster_coeff_of_each_node.append(coeff)
print(cluster_coeff_of_each_node)
```
